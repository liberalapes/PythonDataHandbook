# 4장 Matplotlib을 활용한 시작화

## <u>하나 가격에 인터페이스 두 개</u>

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

### 매트랩 스타일의 인터페이스

```python
x = np.linspace(0,10,100)
```

```python
plt.figure() #플롯 그림 생성
#두 개의 패널 중 첫 번째 패널으 생성하고 현재 축(axis)을 설정
plt.subplot(2,1,1) # (rows, columns, panel number)
plt.plot(x, np.sin(x))

#두 번쨰 패널을 생성하고 현재 축(axis)을 설정
plt.subplot(2,1,2)
plt.plot(x, np.cos(x));
```

### 객체지향 인터페이스

```python
# 먼저 플롯 그리드를 생성
# ax는 두 개의 축 객체의 배열이 됨
fig, ax = plt.subplots(2)

# 적절한 객체에서 plot() 메서드를 호출
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));
```
