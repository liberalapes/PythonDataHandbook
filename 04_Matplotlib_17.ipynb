### Seaborn과 Matplotlib의 차이

```python
import matplotlib.pyplot as plt
plt.style.use('classic')
%matplotlib inline

import numpy as np
import pandas as pd
```

```python
# 데이터 생성
rng = np.random.RandomState(0)
x = np.linspace(0, 10, 500)
y = np.cumsum(rng.randn(500, 6), 0)
```

```python
# Matplotlib 기본값으로 데이터 플롯 그리기
plt.plot(x, y)
plt.legend('ABCDEF', ncol=2, loc='upper left')
```

```python
import seaborn as sns
sns.set()
```

```python
# 위와 동일한 플로팅 코드
plt.plot(x, y)
plt.legend('ABCDEF', ncol=2, loc='upper left')
```

### Seaborn 플롯 탐색하기

#### 히스토그램, KDE, 밀도

```python
data = np.random.multivariate_normal([0,0], [[5,2], [2,2]], size = 2000)
data = pd.DataFrame(data, columns=['x','y'])
for col in 'xy':
    plt.hist(data[col], normed=True, alpha = 0.5)
```

```python
for col in 'xy':
    sns.kdeplot(data[col], shade=True)
```

```python
sns.distplot(data['x'])
sns.distplot(data['y'])
```

```python
sns.kdeplot(data);
```

```python
with sns.axes_style('white'):
    sns.jointplot("x", "y", data, kind='kde');
```

```python
with sns.axes_style('white'):
    sns.jointplot("x", "y", data, kind='hex');
```

#### 페어 플롯

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
sns.pairplot(iris, hue='species', size=2.5);
```

#### 다면 히스토그램

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips['tip_pct'] = 100 * tips['tip'] / tips['total_bill']
grid = sns.FacetGrid(tips, row="sex", col="time", margin_titles = True)
grid.map(plt.hist, 'tip_pct', bins=np.linspace(0, 40, 15));
```

#### 요인 플롯

```python
with sns.axes_style(style='ticks'):
    g = sns.factorplot("day", "total_bill", "sex", data=tips, kind='box')
    g.set_axis_labels("Day", "Total Bill")
```

#### 결합 분포

```python
with sns.axes_style('white'):
    sns.jointplot("total_bill", "tip", data=tips, kind = 'hex')
```

```python
sns.jointplot("total_bill", "tip", data=tips, kind='reg')
```

#### 막대 플롯

```python
planets = sns.load_dataset('planets')
planets.head()
```

```python
with sns.axes_style('white'):
    g = sns.factorplot("year", data=planets, aspect=2, kind="count", color='steelblue')
    g.set_xticklabels(step=5)
```

```python
with sns.axes_style('white'):
    g = sns.factorplot("year", data=planets, aspect=4.0, kind='count',hue='method', order=range(2001, 2015))
    g.set_yticklabels('Number of Planets Discvered')
```

```python
from datetime import timedelta
def convert_time(s):
    h, m, s = map(int, s.split(':'))
    return timedelta(hours=h, minutes=m, seconds=s)

data = pd.read_csv('/Users/chosikc/seaborn-data/marathon-data.csv',  
                   converters={'split':convert_time, 'final':convert_time})
data.head()
```

```python
data.dtypes
```

```python
data['split_sec'] = data['split'].astype(int)/1E9
data['final_sec'] = data['final'].astype(int)/1E9
data.head()
```

```python
with sns.axes_style('white'):
    g = sns.jointplot("split_sec", "final_sec", data, kind='hex')
    g.ax_joint.plot(np.linspace(4000, 16000), 
                    np.linspace(8000, 32000), ':k')
```

```python
data['split_frac'] = 1 - 2 * data['split_sec'] / data['final_sec']
data.head()
```

```python
sns.distplot(data['split_frac'], kde=False);
plt.axvline(0, color="k", linestyle="--");
```

```python
g = sns.PairGrid(data, vars=['age', 'split_sec', 'final_sec', 'split_frac'], hue='gender', palette='RdBu_r')
g.map(plt.scatter, alpha=0.8)
g.add_legend();
```

```python
sns.kdeplot(data.split_frac[data.gender=='M'], label='men', shade=True)
sns.kdeplot(data.split_frac[data.gender=='W'], label='women', shade=True)
plt.xlabel('split_frac');
```

```python
sns.violinplot("gender", "split_frac", data=data, palette=["lightblue", "lightpink"]);
```

```python
data['age_dec'] = data.age.map(lambda age: 10 * (age // 10))
data.head()
```

```python
men = (data.gender == 'M')
women = (data.gender == 'W')

with sns.axes_style(style=None):
    sns.violinplot("age_dec", "split_frac", hue='gender', data=data, 
                   split=True, inner='quartile', palette=["lightblue", "lightpink"]);
```
